```python
import pandas as pd
```

```python
df = pd.read_csv("../data/survey_results_public.csv")

df = df[[
    "Country",
    "YearsCode",
    "DevType",
    "LanguageHaveWorkedWith",
    "ConvertedCompYearly",
    "RemoteWork",
]]

df.info()
```

```python
df['LanguageHaveWorkedWith'].fillna('Unknown', inplace=True)

for col in df.columns:
    if df[col].dtype in ['object', 'category']:
        # категориальные - мода
        if df[col].isna().any():
            df[col] = df[col].fillna(df[col].mode()[0])
    else:
        # числовые - медиана
        if df[col].isna().any():
            df[col] = df[col].fillna(df[col].median())

df.info()
```

```python
df_extra = df.sample(20000, replace=True, random_state=42)

df_final = pd.concat([df, df_extra])

df_final.to_csv("../data/stackoverflow_100k.csv", index=False)

df_final.info()
```
